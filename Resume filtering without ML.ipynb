KMP Algorithm for keyword matching

```python
# KMP Algorithm

def KMP(pat, txt):
	M = len(pat)
	N = len(txt)
	flag=False

	lps = [0]*M
	j = 0 
	computeLPSArray(pat, M, lps)

	i = 0 
	while (N - i) >= (M - j):
		if pat[j] == txt[i]:
			i += 1
			j += 1

		if j == M:
			flag=True
			j = lps[j-1]

		elif i < N and pat[j] != txt[i]:
			if j != 0:
				j = lps[j-1]
			else:
				i += 1
	return(flag)

def computeLPSArray(pat, M, lps):
	len = 0 
	lps[0] = 0 
	i = 1

	while i < M:
		if pat[i] == pat[len]:
			len += 1
			lps[i] = len
			i += 1
		else:
			if len != 0:
				len = lps[len-1]

			else:
				lps[i] = 0
				i += 1

# if __name__=='__main__':
    
# 	txt = "ABABDABACDABABCABAB"
# 	pat = "ABABCABAB"
    
# 	KMP(pat, txt)


```

# Importing modules and files

```python

import openpyxl

excel_file_path = "C:/Users/gowth/Downloads/excelfile1.xlsx"

file=openpyxl.load_workbook(excel_file_path)

filtered_candidates = []
sheet=file.worksheets[0]
```

# Cutoffs

```python
min_py=2
min_ml=1
min_nlp=1
min_dl=1
availabilty = ['Yes, I am available for 3 months']
degree = ['Data Science','Computer Science','B.E','B.Tech','Bachelor of Technology','Artificial Intelligence','Computer','B.Sc','BE','Bachelor of Technology','Data Scientist','M.Sc','M.Tech','MCA','Diploma','M.C.S','MS','M.E']
min_grad_year=2022

desired_keywords_tech = ['Python', 'Machine Learning', 'Natural Language Processing', 'Deep Learning']
desired_keywords_soft = ['English']

min_total_academic_score = 70 #50 percent
```

# Tools

Skill Filtering tool

```python
def tech_skills_filtering(row):
    flag = False
    if row[1] is not None and row[2] is not None and row[3] is not None and row[4] is not None:
        if row[1] >= min_py and row[2] >= min_ml and row[3] >= min_nlp and row[4] >= min_dl:
            flag = True
    return flag
```

Softskill filtering tool

```python
def keyword_matching_tool_soft(row):
    flag = False
    if row[5] is not None:
        for keyword in desired_keywords_soft:
            if KMP(keyword.lower(), row[5].lower()):
                flag = True
    return flag
```

Technical Skills filtering Tool

```python
def keyword_matching_tool_tech(row):
    flag=False
    if row[5] is not None:
        for keyword in desired_keywords_tech:
            if KMP(keyword.lower(),row[5].lower()) == True:
                flag=True
    return flag
```

Availability filtering tool

```python
def keyword_matching_tool_availability(row):
    flag=False
    if row[6] is not None:
        for keyword in availabilty:
            if KMP(keyword.lower(),row[6].lower()) == True:
                flag=True
    return flag
```

Degree Filtering tool

```python
def keyword_matching_tool_degree(row):
    flag=False
    if row[7] is not None:
        for keyword in degree:
            if KMP(keyword.lower(),row[7].lower()) == True:
                flag=True
    return flag
```

Calculating total academic score and making it's filtering tool

```python
from decimal import Decimal
def calculate_academic_score(row_number):
    sheet=file.worksheets[0]
    marks=[]
    for row in sheet[2:1137]:

        if row[row_number].value != 'N/A' or None: 
            value = row[row_number].value.replace(',', '')  # Remove any commas if present
            numerator, denominator = value.split('/')  # Split the string into numerator and denominator
            if Decimal(numerator)!=Decimal(denominator):
                fraction = Decimal(numerator) / Decimal(denominator)
            else:
                fraction= Decimal(numerator)/10
        
            if fraction>1:
                fraction=fraction*10
            elif fraction<0.1:
                fraction=fraction*1000
            else:
                fraction=fraction*100
            marks.append(fraction)
        else:
            marks.append(0)
    return marks

pg_marks=calculate_academic_score(10)
ug_marks=calculate_academic_score(11)
inter_marks=calculate_academic_score(12)
tenth_marks=calculate_academic_score(13)

total_scores=[0]*sheet.max_row

for i in range(len(pg_marks)):
    
    if pg_marks[i]==0:
        total_scores[i]=(ug_marks[i]+inter_marks[i]+tenth_marks[i])/3
    else:
        total_scores[i]=(pg_marks[i]+ug_marks[i]+inter_marks[i]+tenth_marks[i])/4
        




def total_academic_score(x):
    flag=False
    if total_scores[x] is not None and total_scores[x]>= min_total_academic_score:
        flag=True
    return flag

```

# Filtering Process

```python
filtered_candidates = []
row_index=0
for row in sheet.iter_rows(min_row=2, values_only=True):
    
    if total_academic_score(row_index) and tech_skills_filtering(row) and keyword_matching_tool_soft(row) and keyword_matching_tool_tech(row) and keyword_matching_tool_availability(row) and keyword_matching_tool_degree(row):
        filtered_candidates.append(row)
    row_index+=1
```
